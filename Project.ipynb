## Introduction and Background  of the Study

The Newyork City (NYC) yellow taxi is a organization owning several thousands of taxi cabs in Newyork city USA. These taxis are the only vehicles authorized to pick up and transport street-hailing and prearranged passengers anywhere in the city (see https://www.nyc.gov/site/tlc/businesses/yellow-cab.page). The dataset presented in the analysis and modelling carried out in the Report below is one consisting of features that capture information about single taxi trips in Newyork city.The fare amount and other features were provided to the NYC Taxi and Limousine Commission (TLC) by technology providers authorized under the Taxicab & Livery Passenger Enhancement Programs (TPEP/LPEP).

## Tasks/Goals

The aim of this study is to analyse the historical data contained in the New york yellow taxi dataset provided. This analysis effort seeks to find/uncover relationships between the variables contained in the data to ascertain if it is possible to train a predictive model to predict the fare amount charge (target variable) of trips. One real-world application of the results would be to offer actionable insights to everyday users of the yellow taxi service to aid their decision-making.
A second angle to view the upside of results of such analysis is the possible benefits that it offers to the the yellow taxi company. The ability to predict trip charges before-hand given certain factors opens up a new Revenue model that the taxi company could exploit. For example, by offering/upselling this as an extra service in a mobile application

First we visualize how the fare amount  vary with  pick up location. To that we can calculate the distance from the pick-up location using the pick-up latitude and longitude values to the NYC city center

```python
# Gotten from https://stackoverflow.com/questions/27928/
# Function to calculate the distance between two longitude - latitude points using haversine-formula

def distance(lat_1, lon_1, lat_2, lon_2):
    p = 0.017453292519943295 # Pi/180
    a = 0.5 - np.cos((lat_2 - lat_1) * p)/2 + np.cos(lat_1 * p) * np.cos(lat_2 * p) * (1 - np.cos((lon_2 - lon_1) * p)) / 2
    return 0.6213712 * 12742 * np.arcsin(np.sqrt(a))

nyc = (-74.0063889, 40.7141667) # Longitude and Latitude pair for Nyc center

# add new column to dataframe with distance in mile to the city center
taxi_df['dist_center'] = distance(nyc[1], nyc[0], taxi_df.pickup_latitude, taxi_df.pickup_longitude)
```

```python
# Function for visualizing the Rmse, evs and a histogram plot of the prediction errors
def plot_prediction_analysis(y, y_pred, figsize=(10,4), title=''):
    fig, axs = plt.subplots(1, 2, figsize=figsize)
    axs[0].scatter(y, y_pred)
    mn = min(np.min(y), np.min(y_pred))
    mx = max(np.max(y), np.max(y_pred))
    axs[0].plot([mn, mx], [mn, mx], c='red')
    axs[0].set_xlabel('$y$')
    axs[0].set_ylabel('$\hat{y}$')
    rmse = np.sqrt(mean_squared_error(y, y_pred))
    evs = explained_variance_score(y, y_pred)
    axs[0].set_title('rmse = {:.2f}, evs = {:.2f}'.format(rmse, evs))

    axs[1].hist(y-y_pred, bins=50)
    avg = np.mean(y-y_pred)
    std = np.std(y-y_pred)
    axs[1].set_xlabel('$y - \hat{y}$')
    axs[1].set_title('Histrogram prediction error, $\mu$ = {:.2f}, $\sigma$ = {:.2f}'.format(avg, std))

    if title!='':
        fig.suptitle(title)
```

```python
plot_prediction_analysis(y_train, y_train_pred, title='Linear Model - Trainingset')
```

```python
plot_prediction_analysis(y_test, y_test_pred, title='Linear Model - Testset')
```

From the results above, using a baseline linear regression model with the selected features produces a rmse of 2.52 which already denotes a good model for the data, but we can validate our model better using other error metrics.
To achieve this, another good error metric for evaluating our baseline model is the Mean Absolute Percentage error (MAPE) defined by the formula;
M = 1/n E|(At - Ft)/At| 100

where

M = Mean Absolute Percentage Error
n = The number of times the summation occurred
At = Actual value
Ft = Forecast value

MAPE is one of the most popular methods for assessing the performance of machine learning forecasting models.

```python
# Define function for computing MAPE
def mape(Y_actual,Y_Predicted):
    Y_actual = np.array(Y_actual)
    Y_Predicted = np.array(Y_Predicted)
    mape = []
    mape.append((np.abs((Y_actual - Y_Predicted)/Y_actual))*100)
    return mape
```

```python
# Compute the mape on the test set
mape_errors = mape(y_test, y_test_pred)
mape_errors
```

```python
# Plot the errors against time
fig = plt.figure(figsize=(20,10))
plt.hist(mape_errors[0], bins=50)
plt.grid(True)
plt.xlabel('Mape Errors')
plt.ylabel('Occurences')
plt.title('MAPE on the test set using a linear regression model')
ax = plt.gca()
ax.xaxis.set_major_locator(ticker.MultipleLocator(10))
plt.show()
```

Next, we can perform some preliminary feature selection using a linear regression model implementing Ordinary least squares to evaluate the best independent variables that explain the target/dependent variable. OLS compares the difference between individual points in the data set and the predicted best fit line to measure the amount of error produced according to https://medium.com/swlh/interpreting-linear-regression-through-statsmodels-summary-4796d359035a

```python
eval_model = sm.OLS(y, exog= X)
results = eval_model.fit()
print(results.summary())
```

Interpretation of the Results:
1. The R squared metric is a measure of how much of the change in the dependent variable is explained by the independent variables. From our results we can deduce that 88.4% of the fare amount is explained by our selected features
2. The adjusted R squared is metric to penalize the R squared metric based on the number of independent variables as linear regression has the characteristic that the R squared tends to increase or stay the same as additional features are added to the model. From our results, the adjusted R squared is the same as the R squared which indicates that none of the independent variables constitute noise for the model.
3. The Prob (F-statistic) tells us the accuracy of the null hypothesis which is how true the independent variables' effect on our model is zero. From our results, it indicates that there is 0% chance that the effect of the dependent variables is zero which is a good result.
4. P > |t| or p-value tells us how likely the coefficient is measured through the model by chance. Using the coefficient with the largest value 'passenger_count', it tells us that there a 61.7% chance that the passenger count has no effect on the fare amount. This result makes sense because in reality the number of passengers during a ride usually does not affect the amount charged at the end of the trip. This site (https://www.nyc.gov/site/tlc/passengers/taxi-fare.page) corroborates this finding.
